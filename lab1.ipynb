```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
```

```python
uniform_distribution = np.random.uniform(low=0, high=1, size=1000)
normal_distribution_std_dev1 = np.random.normal(loc=0, scale=1, size=1000)
normal_distribution_std_dev_10 = np.random.normal(loc=0, scale=10, size=1000)
beta_distribution_11 = np.random.beta(a=1, b=1, size=1000)
beta_distribution_22 = np.random.beta(a=2, b=2, size=1000)
beta_distribution_39 = np.random.beta(a=3, b=9, size=1000)
two_dim_normal_distribution_1 = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0], [0, 1]], size=10000)
two_dim_normal_distribution_2 = np.random.multivariate_normal(mean=[0, 0], cov=[[2, 1], [1, 2]], size=10000)
```

```python
distributions = [uniform_distribution, normal_distribution_std_dev1, normal_distribution_std_dev_10, beta_distribution_11, beta_distribution_22, beta_distribution_39]
hists_titles = ["Uniform", r'Normal with $\mu = 0, \sigma=1$', r'Normal with $\mu = 0, \sigma=10$', r'Beta with $\alpha = 1, \beta = 1$', r'Beta with $\alpha = 2, \beta = 2$', r'Beta with $\alpha = 3, \beta = 9$']
analytical_formulas = [scipy.stats.uniform, scipy.stats.norm, scipy.stats.norm, scipy.stats.beta, scipy.stats.beta, scipy.stats.beta]
parameters = [(0, 1), (0, 1), (0, 10), (1, 1), (2, 2), (3, 9)]

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)
ax = ax.flatten()
for i, (dist, hist_title, analytical_formula, parameter) in enumerate(zip(distributions, hists_titles, analytical_formulas, parameters)):
  x_values_normal = np.linspace(min(dist), max(dist), 1000)
  param1, param2 = parameter
  #pdf = probability density function
  ax[i].plot(x_values_normal, analytical_formula.pdf(x_values_normal, param1, param2), 'r', label='analytical formula')
  ax[i].hist(dist, density=True, label='hist')
  ax[i].set_title(hist_title)
  ax[i].grid()
plt.suptitle("Histograms for one dimensional distributions")
plt.show()
```

```python
distributions_2d = [two_dim_normal_distribution_1, two_dim_normal_distribution_1, two_dim_normal_distribution_1, two_dim_normal_distribution_2, two_dim_normal_distribution_2, two_dim_normal_distribution_2]
hists2d_titles = ["cov=[[1, 0], [0, 1]], bins=10", "cov=[[2, 1], [1, 2]], bins=20", "cov=[[1, 0], [0, 1]], bins=40", "cov=[[2, 1], [1, 2]], bins=10", "cov=[[1, 0], [0, 1]], bins=20", "cov=[[2, 1], [1, 2]], bins=40"]
analytical_formulas = [scipy.stats.uniform, scipy.stats.norm, scipy.stats.norm, scipy.stats.beta, scipy.stats.beta, scipy.stats.beta]
bins = [10, 20, 40, 10, 20, 40]

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)
ax = ax.flatten()
for i, (dist2d, hist2d_title, bin) in enumerate(zip(distributions_2d, hists2d_titles, bins)):
  ax[i].hist2d(x=dist2d[:, 0], y=dist2d[:, 1], density=True, bins=bin)
  ax[i].set_title(hist2d_title)
  ax[i].grid()
plt.suptitle("Histograms for two dimensional distributions from matplotlib")
plt.show()
```

```python
distributions_2d = [two_dim_normal_distribution_1, two_dim_normal_distribution_1, two_dim_normal_distribution_1, two_dim_normal_distribution_1, two_dim_normal_distribution_1, two_dim_normal_distribution_1]
hists2d_titles = ["cov=[[1, 0], [0, 1]], bins=10", "cov=[[2, 1], [1, 2]], bins=20", "cov=[[1, 0], [0, 1]], bins=40", "cov=[[2, 1], [1, 2]], bins=10", "cov=[[1, 0], [0, 1]], bins=20", "cov=[[2, 1], [1, 2]], bins=40"]
bins = [10, 20, 40, 10, 20, 40]

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)
ax = ax.flatten()
for i, (dist2d, hist2d_title, bin) in enumerate(zip(distributions_2d, hists2d_titles, bins)):
  sns.histplot(x=dist2d[:, 0], y=dist2d[:, 1], ax=ax[i], bins=bin)
  ax[i].set_title(hist2d_title)
  ax[i].grid()
plt.suptitle("Histograms for two dimensional distributions from seaborn")
plt.show()
```

```python
alfa = np.random.normal(loc=2, scale=0.1, size=1000)
beta = np.random.normal(loc=0.2, scale=0.2, size=1000)
x = np.random.normal(loc=0, scale=3, size=1000)
sigma = np.abs(np.random.normal(loc=0, scale=1, size=1000))
mu = alfa + beta * x
y = np.random.normal(loc=mu, scale=sigma, size=1000)
```

```python
x_values = np.linspace(-6, 6, 1000)
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
for i in range(1000):
  y_value = alfa[i] + beta[i] * x_values
  plt.plot(x_values, y_value, 'r')
plt.legend(["y=f(x)", r'$y = \alpha + \beta x$'])
plt.xlim([-6, 6])
plt.grid()
plt.title(r'Comparision between y=f(x) and $y = \alpha + \beta x$')
plt.show()
```
